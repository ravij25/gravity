```python
import pandas as pd
import numpy as np
import pywt
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ravij25/gravity/main/kadapa2.csv")
```

```python
df
```

```python
# Define the level of decomposition
decomp_level = 5
```

```python
# Extract the data as a numpy array
data = df['Bouger anomaly'].to_numpy()
```

```python
# Define the wavelet to be used
wavelet = pywt.Wavelet('db4')
```

```python
# Perform the wavelet decomposition
coeffs = pywt.wavedec(data, wavelet, level=decomp_level)
```

```python
# Reconstruct the data from the wavelet coefficients
reconstructed_data = pywt.waverec(coeffs, wavelet)
```

```python
# Plot the original and reconstructed data for comparison
plt.plot(data, label='Original data')
plt.plot(reconstructed_data, label='Reconstructed data')
plt.legend()
plt.show()
```

```python
# Plot the original and reconstructed data for comparison
plt.figure(figsize=(28, 6))
plt.plot(data, label='Original data')
plt.plot(reconstructed_data, label='Reconstructed data')
plt.legend()
plt.show()
```

```python

# Create a figure with two subplots
fig, axs = plt.subplots(2, 1, figsize=(28, 12))

# Plot the original data on the first subplot
axs[0].plot(data)
axs[0].set_title('Original Data')

# Plot the reconstructed data on the second subplot
axs[1].plot(reconstructed_data)
axs[1].set_title('Reconstructed Data')

# Show the plot
plt.show()
```

```python
# Compute the depth of the source of the Bouguer anomaly
G = 6.67430e-11 # Gravitational constant
density = 2670 # Density of the source material in kg/m^3
volume = 1000 # Volume of the source material in m^3
bouguer_anomaly = df['Bouger anomaly'].mean() # Mean Bouguer anomaly in mGal

depth = (2 * np.pi * G * density * volume) / (1000 * bouguer_anomaly)

print("Depth of the source of the Bouguer anomaly: {:.2f} meters".format(depth))

# Plot the original and reconstructed data for comparison
plt.figure(figsize=(28, 6))
plt.plot(data, label='Original data')
plt.plot(reconstructed_data, label='Reconstructed data')
plt.legend()
plt.show()
```
