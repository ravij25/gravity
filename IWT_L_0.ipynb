```python
import numpy as np
import pandas as pd
import pywt
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ravij25/gravity/main/wavelet_coeffs_level0.csv")
```

```python
df
```

```python
signal = df['Level 0'].to_numpy()
```

```python
wavelet = 'db4'
decomp_level = 5
```

```python
coeffs = pywt.wavedec(signal, wavelet, level=decomp_level)
```

```python
reconstructed_signal = pywt.waverec(coeffs, wavelet)
```

```python
df_reconstructed = pd.DataFrame({'your_column': reconstructed_signal})
```

```python
df_reconstructed.to_csv('reconstructed_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Plot the original and reconstructed signals
fig, axs = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
axs[0].plot(signal)
axs[0].set_title('Original Signal')
axs[1].plot(reconstructed_signal)
axs[1].set_title('Reconstructed Signal')
plt.tight_layout()
plt.show()
```
